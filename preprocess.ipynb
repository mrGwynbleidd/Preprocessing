```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Dataset

```python
df = pd.read_csv("drawndata1.csv")
```

# Review data

```python
df.head(3)
```

```python
X = df[['x', 'y']].values
y = df['z'] == "a"
```

```python
plt.scatter(X[:,0], X[:,1], c = y)
```

# scaling to get simillar numerical values on x and y axis

```python
from sklearn.preprocessing import StandardScaler
X_new = StandardScaler().fit_transform(X)
plt.scatter(X_new[:,0], X_new[:,1], c = y)
#but we still have outlayers
```

# Remove outlayers

```python
from sklearn.preprocessing import QuantileTransformer
X_new = QuantileTransformer(n_quantiles=100).fit_transform(X)
plt.scatter(X_new[:,0], X_new[:,1], c = y)
```

# Create Pipeline

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline 

def plot_output(scaler):
    pipe = Pipeline([
        ("scale", scaler),
        ("model", KNeighborsClassifier(n_neighbors=20, weights='distance'))
    ])

    pred = pipe.fit(X, y).predict(X)

    plt.figure(figsize=(9, 3))
    plt.subplot(131)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.title("Original Data")
    plt.subplot(132)
    X_tfm = scaler.transform(X)
    plt.scatter(X_tfm[:, 0], X_tfm[:, 1], c=y)
    plt.title("Transformed Data")
    plt.subplot(133)
    X_new = np.concatenate([
        np.random.uniform(0, X[:, 0].max(), (5000, 1)), 
        np.random.uniform(0, X[:, 1].max(), (5000, 1))
    ], axis=1)
    y_proba = pipe.predict_proba(X_new)
    plt.scatter(X_new[:, 0], X_new[:, 1], c=y_proba[:, 1], alpha=0.7)
    plt.title("Predicted Data")
```

# Review Transformed Data

```python
plot_output(scaler=StandardScaler())
```

```python
plot_output(scaler=QuantileTransformer(n_quantiles=100))
```

# Load dataset, where it is not seperable

```python
df = pd.read_csv("drawndata2.csv")
X = df[['x', 'y']].values
y = df['z'] == 'a'
plt.scatter(X[:, 0], X[:, 1], c=y);
```

# Logistic is bad classifier 

```python
from sklearn.linear_model import LogisticRegression

from sklearn.pipeline import Pipeline

pipe = Pipeline([
    #preprocessing step
    ("scale", QuantileTransformer(n_quantiles=100)),
    ("model", LogisticRegression())
])

pred = pipe.fit(X, y).predict(X)
plt.scatter(X[:, 0], X[:, 1], c=pred);
```

# Convert to nonlinear

```python
from sklearn.preprocessing import PolynomialFeatures

pipe = Pipeline([
    #preprocessing step
    ("scale", PolynomialFeatures()),
    ("model", LogisticRegression())
])

pred = pipe.fit(X, y).predict(X)
plt.scatter(X[:, 0], X[:, 1], c=pred);
```
